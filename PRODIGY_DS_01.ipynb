```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'Salarydata.csv'
salaries = pd.read_csv(file_path)

a = salaries.Department.value_counts()
x = list(a.index)
y = list(a)

f, ax = plt.subplots(figsize=(20,10))
width = 0.75
ind = np.arange(len(y))
ax.barh(ind, y, width, color="black")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(x, minor=False)
for i, v in enumerate(y):
    ax.text(v + .25, i + .25, str(v), color='red', fontweight='bold')
plt.title('Departmentwise Avg Pay')
plt.xlabel('Avg Base Pay')
plt.ylabel('Dept')
plt.show()
```
